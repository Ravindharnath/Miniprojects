```python
df_us_counties_recent = pd.read_csv("/content/drive/MyDrive/us-counties (1).csv")
df_us_counties = pd.read_csv("/content/drive/MyDrive/us-counties (1).csv")
df_us_counties_recent.head(5)
df_us_counties.head(5)
```

```python
df3 = pd.merge(df_us_counties_recent, df_us_counties, how='outer')
df3.head( 10)
```

```python
# 1.Top 10 states having highest total death in the year 2020
def get_year(row):
    return datetime.strptime(row['date'], '%Y-%m-%d').year # Use datetime.strptime directly

df3['year'] = df3.apply(get_year, axis=1)
df_2020 = df3.where(df3['year'] == 2020)
df_top10_2020 = df_2020.groupby(['state'])['deaths'].sum().to_frame('count').reset_index()
df_top10_2020 = df_top10_2020.sort_values(by=['count'], ascending=False)
df_top10_2020.head(10)
```

```python
# 2.Top 10 states having highest total death in the year 2021
def get_year(row):
    return datetime.strptime(row['date'], '%Y-%m-%d').year # Use datetime.strptime directly

df3['year'] = df3.apply(get_year, axis=1)
df_2021 = df3.where(df3['year'] == 2021)
df_top10_2021 = df_2021.groupby(['state'])['deaths'].sum().to_frame('count').reset_index()
df_top10_2021 = df_top10_2021.sort_values(by=['count'], ascending=False)
df_top10_2021.head(10)
```

```python
# 3.Top 10 states having highest average death in the year 2020
def get_year(row):
    return datetime.strptime(row['date'], '%Y-%m-%d').year # Use datetime.strptime directly

df3['year'] = df3.apply(get_year, axis=1)
df_2020 = df3.where(df3['year'] == 2020)
df_top10_2020 = df_2020.groupby(['state'])['deaths'].mean().to_frame('count').reset_index()
df_top10_2020 = df_top10_2020.sort_values(by=['count'], ascending=False)
df_top10_2020.head(10)
```
